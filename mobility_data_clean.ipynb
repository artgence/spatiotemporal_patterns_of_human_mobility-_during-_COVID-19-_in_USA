```python
import pandas as pd
import dask.dataframe as dd
```

```python
df = dd.read_csv('2020_US_Region_Mobility_Report.csv',dtype={'grocery_and_pharmacy_percent_change_from_baseline': 'float64',
       'iso_3166_2_code': 'object',
       'parks_percent_change_from_baseline': 'float64',
       'residential_percent_change_from_baseline': 'float64',
       'retail_and_recreation_percent_change_from_baseline': 'float64',
       'sub_region_1': 'object',
       'sub_region_2': 'object',
       'transit_stations_percent_change_from_baseline': 'float64',
       'workplaces_percent_change_from_baseline': 'float64'})
df = df.dropna(subset=['iso_3166_2_code'])
df = df[['sub_region_1','date','retail_and_recreation_percent_change_from_baseline','grocery_and_pharmacy_percent_change_from_baseline', 
         'parks_percent_change_from_baseline','transit_stations_percent_change_from_baseline',
         'workplaces_percent_change_from_baseline','residential_percent_change_from_baseline']].compute()
df.head()
```

```python
df1 = dd.read_csv('2021_US_Region_Mobility_Report.csv',dtype={'grocery_and_pharmacy_percent_change_from_baseline': 'float64',
       'iso_3166_2_code': 'object',
       'parks_percent_change_from_baseline': 'float64',
       'residential_percent_change_from_baseline': 'float64',
       'retail_and_recreation_percent_change_from_baseline': 'float64',
       'sub_region_1': 'object',
       'sub_region_2': 'object',
       'transit_stations_percent_change_from_baseline': 'float64',
       'workplaces_percent_change_from_baseline': 'float64'})
df1 = df1.dropna(subset=['iso_3166_2_code'])
df1 = df1[['sub_region_1','date','retail_and_recreation_percent_change_from_baseline','grocery_and_pharmacy_percent_change_from_baseline', 
         'parks_percent_change_from_baseline','transit_stations_percent_change_from_baseline',
         'workplaces_percent_change_from_baseline','residential_percent_change_from_baseline']].compute()
df1.head()
```

```python
df2 = dd.read_csv('2022_US_Region_Mobility_Report.csv',dtype={'grocery_and_pharmacy_percent_change_from_baseline': 'float64',
       'iso_3166_2_code': 'object',
       'parks_percent_change_from_baseline': 'float64',
       'residential_percent_change_from_baseline': 'float64',
       'retail_and_recreation_percent_change_from_baseline': 'float64',
       'sub_region_1': 'object',
       'sub_region_2': 'object',
       'transit_stations_percent_change_from_baseline': 'float64',
       'workplaces_percent_change_from_baseline': 'float64'})
df2 = df2.dropna(subset=['iso_3166_2_code'])
df2 = df2[['sub_region_1','date','retail_and_recreation_percent_change_from_baseline','grocery_and_pharmacy_percent_change_from_baseline', 
         'parks_percent_change_from_baseline','transit_stations_percent_change_from_baseline',
         'workplaces_percent_change_from_baseline','residential_percent_change_from_baseline']].compute()
df2.head()
```

```python
df_all = pd.concat([df,df1,df2],ignore_index=True)
new_order = [1,0,2,3,4,5,6,7]
df_all = df_all[df_all.columns[new_order]]
df_all = df_all.sort_values(by=['date','sub_region_1'])
df_all = df_all.fillna(0)
df_all.head()
```

```python
df_all.to_csv('US_Region_Mobility_Report_cleaned.csv',index=False)
```
