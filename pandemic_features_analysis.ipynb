```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pandemic_data = pd.read_csv("covid_df_cleaned.csv")
```

```python
pandemic_data.drop(columns=pandemic_data.columns[0], axis=1, inplace=True)
```

```python
# Correlation of COVID-19 pandemic indicators
pandemic_data_corr = pandemic_data[['Confirmed', 'Pop', 
                               'Lag_Confirmed_2', 'Lag_Confirmed_5', 
                               'Lag_Deaths_2', 'Lag_Deaths_5', 
                               'Lag_Recovered_2', 'Lag_Recovered_5', 
                               'Trend_Confirmed_2', 'Trend_Confirmed_5', 
                               'Trend_Deaths_2', 'Trend_Deaths_5', 
                               'Trend_Recovered_2', 'Trend_Recovered_5', 
                               'Deaths_ema', 'mortality_factor', 'mortality_rate']]

ax = sns.heatmap(
    pandemic_data_corr.corr(), 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
)
ax.set_title('Correlation of Basic COVID-19 Metrics')
ax
```

```python
def heatmap_numeric_w_dependent_variable(df, dependent_variable):
    '''
    Takes df, a dependant variable as str
    Returns a heatmap of all independent variables' correlations with dependent variable 
    '''
    plt.figure(figsize=(8, 10))
    g = sns.heatmap(df.corr()[[dependent_variable]].sort_values(by=dependent_variable), 
                    annot=True, 
                    cmap='coolwarm', 
                    vmin=-1,
                    vmax=1) 
    return g
```

```python
pandemic_data_vari = pandemic_data.drop(columns=['Province_State', 'Covid_Date', 'Day', 'Month', 'Year'])
heatmap_numeric_w_dependent_variable(pandemic_data_vari, 'Confirmed')
```

```python
heatmap_numeric_w_dependent_variable(pandemic_data_vari, 'Deaths')
```

```python
heatmap_numeric_w_dependent_variable(pandemic_data_vari, 'Recovered')
```
